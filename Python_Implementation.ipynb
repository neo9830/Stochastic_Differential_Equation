### The following implementation simulates the location and Speed of an object in a continous system using the Euler Maruyama method with Ito Interal used for the simulation. 
### The total time of the continous process is configured to be 10 units. And the Drift and Diffusion coefficient for the equation is configured to be 0.1 and 0.05 respectively

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T = 10.0      # Total time
N = 1000      # Number of time steps
dt = T / N    # Time step
mu = 0.1      # Drift coefficient for speed
sigma = 0.05   # Diffusion coefficient for speed

# Time array which is total for 10 seconds but with 1000 steps
t = np.linspace(0, T, N+1)

V = np.zeros(N+1) #Speed Array
X = np.zeros(N+1) #Position Array

# Initializing the Speed and Position value
V[0] = 0     # Initial speed
X[0] = 0     # Initial position

# Process Simulation
for i in range(1, N+1):
    dW = np.sqrt(dt) * np.random.randn()  # Increment of Wiener process
    V[i] = V[i-1] + mu * dt + sigma * dW  # Update speed using Euler-Maruyama
    X[i] = X[i-1] + V[i-1] * dt           # Update position by integrating speed

# Plot the results
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(t, V, label='Speed $V_t$')
plt.xlabel('Time $t$')
plt.ylabel('Speed $V_t$')
plt.title('Speed of the Car Over Time')
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(t, X, label='Position $X_t$', color='orange')
plt.xlabel('Time $t$')
plt.ylabel('Position $X_t$')
plt.title('Position of the Car Over Time')
plt.legend()

plt.tight_layout()
plt.show()
```

### Now the simulation to be more accurate, it is required to have large sets of data points. In our case it is important to have more time steps for the continous system so that the simulation can be acheived with more accuracy rate. To support the idea, Monte carlo implementation is explained in the below program to calcuate

```python
import random
import math

# Function to estimate Pi using Monte Carlo method
def monte_carlo_pi(num_samples):
    inside_circle = 0
    
    for _ in range(num_samples):
        # Generate random (x, y) points in the range [0, 1]
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        
        # Check if the point is inside the quarter circle (x^2 + y^2 <= 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
    
    # Estimate Pi as 4 * (number of points inside circle / total points)
    pi_estimate = 4 * (inside_circle / num_samples)
    return pi_estimate

# Run the simulation with 1,000,000 samples
num_samples = 1000000
pi_estimate = monte_carlo_pi(num_samples)

print(f"Estimated Pi after {num_samples} samples: {pi_estimate}")
print(f"Actual Pi value: {math.pi}")
  
```
